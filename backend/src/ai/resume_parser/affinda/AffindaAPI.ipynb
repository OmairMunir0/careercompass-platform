```python
from pathlib import Path
from affinda import AffindaAPI, TokenCredential

API_KEY = "API_KEY"
WORKSPACE_ID = "ID"
FILE_PATH = Path("/content/Cv Resume.pdf")

# Authenticate
credential = TokenCredential(token=API_KEY)
client = AffindaAPI(credential=credential)

# Upload and parse resume
with FILE_PATH.open("rb") as f:
    resume = client.create_resume(
        file=f,
    )

# Access parsed data
print(resume.data)
```

```python
resume_data = resume.data  # shortcut

def safe_get(obj, attr):
    """Helper: safely get attribute, return None if missing"""
    return getattr(obj, attr, None)

# === BASIC INFO ===
print("👤 PERSONAL DETAILS")
print("Name:", resume_data.name)
print("Email(s):", ", ".join(resume_data.emails or []))
print("Phone(s):", ", ".join(resume_data.phone_numbers or []))
print("LinkedIn:", resume_data.linkedin or "N/A")
print("Website(s):", ", ".join(resume_data.websites or []))
if resume_data.location:
    print("Location:", safe_get(resume_data.location, "formatted") or "N/A")
print("-" * 60)

# === SUMMARY ===
if resume_data.summary:
    print("📝 SUMMARY")
    print(resume_data.summary)
    print("-" * 60)

# === EDUCATION ===
if resume_data.education:
    print("🎓 EDUCATION")
    for edu in resume_data.education:
        org = safe_get(edu, "organization") or "Unknown Institute"
        degree = safe_get(edu, "accreditation") or safe_get(edu, "accreditation_name") or ""
        start = safe_get(edu, "date_start") or safe_get(edu, "start_date") or ""
        end = safe_get(edu, "date_end") or safe_get(edu, "end_date") or ""
        grade = safe_get(edu, "grade") or ""
        print(f"{org} | {degree}")
        print(f"{grade}  ({start} – {end})\n")
    print("-" * 60)

# === EXPERIENCE ===
if resume_data.work_experience:
    print("💼 WORK EXPERIENCE")
    for exp in resume_data.work_experience:
        title = safe_get(exp, "job_title") or "Unknown Role"
        company = safe_get(exp, "organisation") or "Unknown Company"
        start = safe_get(exp, "date_start") or safe_get(exp, "start_date") or ""
        end = safe_get(exp, "date_end") or safe_get(exp, "end_date") or ""
        print(f"{title} at {company}")
        print(f"Duration: {start} – {end}")
        if safe_get(exp, "location"):
            print("Location:", safe_get(exp.location, "formatted") or "N/A")
        if safe_get(exp, "bullets"):
            for bullet in exp.bullets:
                print(f"  • {bullet}")
        elif safe_get(exp, "summary"):
            print("  •", exp.summary)
        print()
    print("-" * 60)

# === SKILLS ===
if resume_data.skills:
    print("🧠 SKILLS")
    for skill in resume_data.skills:
        print(f"- {safe_get(skill, 'name')}")
    print("-" * 60)

# === PROJECTS (from raw text fallback) ===
text = safe_get(resume_data, "raw_text") or ""
if "Top Projects" in text:
    print("💡 TOP PROJECTS (From Resume Text)")
    project_section = text.split("Top Projects")[-1].split("Technical Skills")[0]
    print(project_section.strip())
print("-" * 60)
```
