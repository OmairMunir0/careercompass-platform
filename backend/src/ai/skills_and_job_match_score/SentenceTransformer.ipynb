```python
candidates_json = [
  {
    "candidate_id": "C001",
    "name": "Ayesha Khan",
    "skills": ["Python", "Machine Learning", "Data Analysis", "Pandas", "SQL"],
    "education": "BS Computer Science",
    "experience": ["Data Analyst Intern", "Machine Learning Research Assistant"]
  },
  {
    "candidate_id": "C002",
    "name": "Ali Raza",
    "skills": ["React", "Node.js", "Express", "JavaScript", "MongoDB"],
    "education": "BS Software Engineering",
    "experience": ["Frontend Developer Intern", "Full Stack Trainee"]
  },
  {
    "candidate_id": "C003",
    "name": "Sara Ahmed",
    "skills": ["UI/UX Design", "Figma", "Adobe XD", "Prototyping", "User Research"],
    "education": "BS Design and Multimedia",
    "experience": ["UI Designer Intern", "Graphic Designer"]
  },
  {
    "candidate_id": "C004",
    "name": "Hamza Tariq",
    "skills": ["C++", "CUDA", "OpenMP", "Parallel Programming"],
    "education": "BS Computer Engineering",
    "experience": ["High-Performance Computing Project", "Teaching Assistant"]
  }
]

recruiter_json = [
  {
    "job_id": "J101",
    "title": "Machine Learning Intern",
    "description": "Assist in developing predictive models and data pipelines using Python and ML frameworks.",
    "required_skills": ["Python", "Machine Learning", "Data Science"]
  },
  {
    "job_id": "J102",
    "title": "Full Stack Developer",
    "description": "Work on React and Node.js-based applications to develop scalable web solutions.",
    "required_skills": ["React", "Node.js", "JavaScript", "Express"]
  },
  {
    "job_id": "J103",
    "title": "UI/UX Designer",
    "description": "Create user-friendly designs and prototypes for mobile and web applications.",
    "required_skills": ["UI/UX Design", "Figma", "Prototyping"]
  },
  {
    "job_id": "J104",
    "title": "Parallel Systems Intern",
    "description": "Work on GPU-based optimization and CUDA kernel design for high-performance systems.",
    "required_skills": ["C++", "CUDA", "OpenMP"]
  }
]
```

```python
from sentence_transformers import SentenceTransformer, util
import json
```

```python
model = SentenceTransformer('all-mpnet-base-v2')
```

```python
candidates = candidates_json
jobs_internships = recruiter_json

print(candidates[0])
print(jobs_internships[0])
```

```python
def textify_candidate(c):
    return " ".join(c["skills"]) + " " + c["education"] + " " + " ".join(c["experience"])

def textify_job(j):
    return j["title"] + " " + " ".join(j["required_skills"]) + " " + j["description"]
```

```python
for c in candidates:
    for j in jobs_internships:
        sim = util.cos_sim(
            model.encode(textify_candidate(c)),
            model.encode(textify_job(j))
        )
        sim_score = float(sim)

        if sim_score >= 0.85:
            label = 4
            desc = "Excellent match"
        elif sim_score >= 0.75:
            label = 3
            desc = "Good match"
        elif sim_score >= 0.6:
            label = 2
            desc = "Moderate match"
        elif sim_score >= 0.5:
            label = 1
            desc = "Low match"
        else:
            label = 0
            desc = "Poor match"

        print(f"{c['candidate_id']} - {j['job_id']} | similarity: {sim_score:.2f} | match_label: {label} ({desc})")
```
